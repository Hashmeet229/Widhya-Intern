```python
chahar_stats= deliveries_data[['match_id','bowler','batting_team','total_runs','ball','over','player_dismissed']]
chahar_stats=chahar_stats[(chahar_stats['batting_team'] == 'Kolkata Knight Riders') & (chahar_stats['bowler'] == 'DL Chahar')]
chahar_stats=chahar_stats.fillna(0)
chahar_stats['player_dismissed'] = chahar_stats['player_dismissed'].apply(lambda x: 1  if x!= 0 else 0)
chahar_stats
```

```python
import statistics as st
balls=[]#store the no. of balls until the first wicket occurs
ids=chahar_stats['match_id'].unique()
for id in ids:
    ball=0
    for i in range(len(chahar_stats)):
        if(chahar_stats['match_id'].iloc[i]==id):
                ball+=1
                if(chahar_stats['player_dismissed'].iloc[i]==1):
                    balls.append(ball)
                    break
                else:
                    continue
                
                
print(balls)
print("MEAN:",st.mean(balls))
print("MEDIAN:",st.median(balls))
```
