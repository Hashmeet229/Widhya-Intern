# QUES 1

```python
df2=df1[['id','team1','team2','winner','result','toss_winner','toss_decision','season']]
df2.head()
```

```python
df3 = df2[(df2['team1']=='Chennai Super Kings') & (df2['team2'] == 'Rajasthan Royals') ]

df3 
```

```python
df4 = df2[(df2['team1']=='Rajasthan Royals') & (df2['team2'] == 'Chennai Super Kings')]

df4 
```

```python
df5 = pd.concat([df3,df4], axis=0)
df5
```

```python
df5['winner'].value_counts()
```

```python
matches=df5['id'].nunique()
matches
```

```python
df5['winner'].value_counts(normalize=True)
```

```python
matches_played=pd.concat([df1['team1'],df1['team2']])
matches_played=matches_played.value_counts().reset_index()
matches_played.columns=['Team','Total Matches']
matches_played['wins']=df1['winner'].value_counts().reset_index()['winner']

matches_played.set_index('Team',inplace=True)
matches_played.reset_index().head(8)
```

```python
win_percentage = round(matches_played['wins']/matches_played['Total Matches'],3)*100
win_percentage.head(8)
```

# QUES 2

```python
df6= df[['match_id','bowling_team','batting_team','over','total_runs']]
df6=df6[(df6['over'] < 7)]
df6

```

```python
df7=df6[(df6['bowling_team']=='Chennai Super Kings') & (df6['batting_team'] == 'Rajasthan Royals')]
df7
```

```python
df8=df7.groupby('batting_team').agg({'total_runs':'sum'})
df8
```

```python
total_runs_in_firstsix_overs=df7['total_runs'].sum()
```

```python
total_runs_in_firstsix_overs_per_match_for_RR=total_runs_in_firstsix_overs/matches
total_runs_in_firstsix_overs_per_match_for_RR
```

# QUES 3

```python
df9= df[['match_id','bowler','batting_team','bye_runs','legbye_runs','total_runs','ball']]
df10=df9[(df9['batting_team']=='Rajasthan Royals') & (df9['bowler'] == 'DL Chahar')]
df10
```

```python
df10['runs against bowler'] = df10['total_runs'] - (df10['bye_runs'] + df10['legbye_runs'])
df10
```

```python
runs_against_chahar=df10['runs against bowler'].sum()
runs_against_chahar
```

```python
matches_for_chahar_against_RR=df10['match_id'].nunique()
matches_for_chahar_against_RR
```

```python
total_no_of_balls_bowled=df10['ball'].value_counts().sum()
total_no_of_balls_bowled
```

```python
economy_rate=(runs_against_chahar/total_no_of_balls_bowled)*6
economy_rate
```
